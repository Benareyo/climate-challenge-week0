```python
import pandas as pd
import numpy as np

# 1. Load dataset
df = pd.read_csv("ethiopia.csv")

# 2. Add Country column
df["Country"] = "Ethiopia"

# 3. Convert YEAR + DOY → real date
df["date"] = pd.to_datetime(df["YEAR"] * 1000 + df["DOY"], format="%Y%j")

# 4. Extract Month
df["Month"] = df["date"].dt.month

# 5. Preview
df.head()
```

```python
# 1. Replace NASA missing values (-999) with real NaN
df = df.replace(-999, np.nan)

# 2. Check duplicates
duplicates = df.duplicated().sum()
print("Number of duplicate rows:", duplicates)

# 3. Drop duplicates
df = df.drop_duplicates()

# 4. Missing values count
missing = df.isna().sum()
print("\nMissing values per column:\n", missing)

# 5. Percentage of missing values
missing_percent = (df.isna().sum() / len(df)) * 100
print("\nMissing percentage:\n", missing_percent)
```

```python
import numpy as np

# columns to check
cols = ["T2M", "T2M_MAX", "T2M_MIN", "PRECTOTCORR", "RH2M", "WS2M", "WS2M_MAX"]

# compute z-scores manually
z_scores = (df[cols] - df[cols].mean()) / df[cols].std()

# find outliers where |z| > 3
outliers = (np.abs(z_scores) > 3)

# count total outlier rows
outlier_count = outliers.any(axis=1).sum()

print("Number of outlier rows:", outlier_count)
```

```python
# We keep outliers because extreme climate events are important for analysis
print("Decision: Keeping outliers for climate extreme analysis")
```

```python

import matplotlib.pyplot as plt

# -------------------------------
# STEP 1: Ensure Month exists
# -------------------------------
df["date"] = pd.to_datetime(df["YEAR"] * 1000 + df["DOY"], format="%Y%j")
df["Month"] = df["date"].dt.month

# -------------------------------
# STEP 2: Monthly Temperature
# -------------------------------
monthly_temp = df.groupby("Month")["T2M"].mean()

plt.figure()
plt.plot(monthly_temp.index, monthly_temp.values, marker="o")
plt.title("Monthly Average Temperature (Ethiopia)")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.grid()
plt.show()

# -------------------------------
# STEP 3: Monthly Rainfall
# -------------------------------
monthly_rain = df.groupby("Month")["PRECTOTCORR"].sum()

plt.figure()
plt.bar(monthly_rain.index, monthly_rain.values)
plt.title("Monthly Total Rainfall (Ethiopia)")
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# select only numeric columns
numeric_df = df[["T2M", "T2M_MAX", "T2M_MIN", "T2M_RANGE",
                 "PRECTOTCORR", "RH2M", "WS2M", "WS2M_MAX"]]

corr = numeric_df.corr()

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Ethiopia Climate Data)")
plt.show()
```
